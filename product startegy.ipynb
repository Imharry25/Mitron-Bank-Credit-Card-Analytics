```python
# Age distribution
plt.figure(figsize=(6,4))
sns.countplot(data=customers, x='age_group')
plt.title('Customer Count by Age Group')
plt.show()
```

```python
# Occupation distribution
plt.figure(figsize=(6,4))
sns.countplot(data=customers, y='occupation', order=customers['occupation'].value_counts().index)
plt.title('Customer Count by Occupation')
plt.show()
```

```python
#Gender Distribution
plt.figure(figsize=(3,2))
sns.countplot(data=customers, x='gender')
plt.title("Customer count by gender")
plt.show()
```

```python
#City Distribution
plt.figure(figsize=(6,4))
sns.countplot(data=customers, x="city")
plt.title('Customer count by city')
plt.show()
```

```python
#B Spend Categories
#Top spending Categories
spends['category'].value_counts().plot(kind='bar')
plt.title("Transaction vol by spend category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
#Payment type
sns.countplot(data=spends, x='payment_type',order=spends['payment_type'].value_counts().index)
plt.title('Transaction count by payment type')
plt.show()
```

```python
#Bivariate & Multivariate Analysis
# A. Merge data for Segment Analysis
     # Aggregate total spend per customer

customer_spend = spends.groupby('customer_id')['spend'].sum().reset_index()
df = pd.merge(customers, customer_spend, on='customer_id', how='left').fillna(0)
```

```python
# Calcualte income Utilisation %
df['income_utilisation_pct'] = df['spend']/df['avg_income']*100
```

```python
# B. Income Utilisation by Segment
```

```python
# By Occupation
plt.figure(figsize=(6,4))
df.groupby('occupation')['income_utilisation_pct'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title('Avg Income Utilisation % by occupation')
plt.ylabel('Avg Utilisation %')
plt.show()
```

```python
# By City
plt.figure(figsize=(6,4))
df.groupby('city')['income_utilisation_pct'].mean().sort_values(ascending=False).plot(kind='bar')
plt.title("Avg Income Utilization % by City")
plt.ylabel("Average Utilization (%)")
plt.show()
```

```python
# By Age group and gender
plt.figure(figsize=(6,4))
pivot = df.pivot_table(index='age_group', columns = 'gender', values='income_utilisation_pct',aggfunc='mean')
pivot.plot(kind='bar')
plt.title('Avg Income Utilization % by Age Group and Gender')
plt.ylabel('Average Utilization (%)')
plt.show()
```

```python
# Merge spends with customers to add city info
spends_with_city = pd.merge(
    spends, customers[['customer_id', 'city']], 
    on='customer_id', how='left'
)
```

```python
df
```

```python
# C. Spend by Categories and City
spending_by_cat_city = spends_with_city.groupby(['city','category'])['spend'].sum().unstack().fillna(0)
spending_by_cat_city.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Total Spend by Category and City')
plt.ylabel('Spend')
plt.show()
```
